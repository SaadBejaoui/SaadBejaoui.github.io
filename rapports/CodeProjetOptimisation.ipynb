```python
# ============================================================================
# MINI-PROJET D'OPTIMISATION - ENIT 2025/2026
# CODE COMPLET AVEC TOUTES LES VISUALISATIONS
# ============================================================================
# Auteur : Bejaoui Saad Eddine
# Classe : 2ème Année MIndS
# Année universitaire : 2025/2026
# ============================================================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import time
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')

print("=" * 70)
print("MINI-PROJET D'OPTIMISATION - ENIT 2025/2026")
print("=" * 70)

# ============================================================================
# PARTIE I: OPTIMISATION SANS CONTRAINTES
# ============================================================================

def partie_I():
    """Exécute toute la Partie I avec toutes les visualisations"""
    print("\n" + "=" * 70)
    print("PARTIE I: OPTIMISATION SANS CONTRAINTES")
    print("=" * 70)

    # ============================================================================
    # FONCTIONS UTILITAIRES
    # ============================================================================

    def construire_matrice_A(n):
        """Construit la matrice de rigidité A (tridiagonale)"""
        h = 1.0 / (n + 1)
        main_diag = 2 * np.ones(n)
        off_diag = -1 * np.ones(n-1)
        A = (n+1)**2 * (np.diag(main_diag) + np.diag(off_diag, -1) + np.diag(off_diag, 1))
        return A, h

    def construire_vecteur_b(n, f_type=1):
        """Construit le vecteur b selon f(x)=1 ou f(x)=π²sin(πx)"""
        h = 1.0 / (n + 1)
        x = np.linspace(0, 1, n+2)[1:-1]

        if f_type == 1:
            f_vals = np.ones(n)
        else:  # f_type == 2
            f_vals = np.pi**2 * np.sin(np.pi * x)

        b = h * f_vals
        return b, x

    def J_n(u, A, b):
        """Fonctionnelle J_n(u) = 0.5*u^T A u - b^T u"""
        return 0.5 * np.dot(u, np.dot(A, u)) - np.dot(b, u)

    def DJ_n(u, A, b):
        """Gradient ∇J_n(u) = A u - b"""
        return np.dot(A, u) - b

    # ============================================================================
    # QUESTION 4: Visualisation pour n=2
    # ============================================================================
    print("\n--- Question 4: Visualisation pour n=2 ---")

    n = 2
    A, h = construire_matrice_A(n)
    b, x = construire_vecteur_b(n, f_type=1)

    # Visualisation 3D complète
    u1_vals = np.linspace(-10, 10, 100)
    u2_vals = np.linspace(-10, 10, 100)
    U1, U2 = np.meshgrid(u1_vals, u2_vals)
    Z = np.zeros_like(U1)

    for i in range(U1.shape[0]):
        for j in range(U1.shape[1]):
            u = np.array([U1[i, j], U2[i, j]])
            Z[i, j] = J_n(u, A, b)

    # Figure 1: Surface 3D et courbes de niveau
    fig = plt.figure(figsize=(14, 6))

    # Surface 3D
    ax1 = fig.add_subplot(121, projection='3d')
    surf = ax1.plot_surface(U1, U2, Z, cmap='viridis', alpha=0.8, rstride=5, cstride=5)
    ax1.set_xlabel('$u_1$', fontsize=12, labelpad=10)
    ax1.set_ylabel('$u_2$', fontsize=12, labelpad=10)
    ax1.set_zlabel('$J_2(u)$', fontsize=12, labelpad=10)
    ax1.set_title('Surface de $J_2$ pour $n=2$', fontsize=14, pad=15)
    fig.colorbar(surf, ax=ax1, shrink=0.6)

    # Courbes de niveau
    ax2 = fig.add_subplot(122)
    contour = ax2.contour(U1, U2, Z, levels=30, cmap='viridis', linewidths=1.5)
    ax2.clabel(contour, inline=True, fontsize=8)
    ax2.set_xlabel('$u_1$', fontsize=12)
    ax2.set_ylabel('$u_2$', fontsize=12)
    ax2.set_title('Courbes de niveau de $J_2$', fontsize=14)
    ax2.grid(True, alpha=0.3)
    ax2.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax2.axvline(x=0, color='k', linestyle='-', alpha=0.3)

    plt.suptitle('Question 4: Visualisation de $J_2$ pour $n=2$', fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('figure1_J2_visualisation.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ============================================================================
    # QUESTION 5: Propriétés de la matrice A
    # ============================================================================
    print("\n--- Question 5: Propriétés de la matrice A ---")

    # Analyse spectrale
    dimensions = [2, 5, 10, 20, 50]
    donnees_spectrales = []

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()

    for idx, n_test in enumerate(dimensions):
        A_test, h_test = construire_matrice_A(n_test)
        eigenvalues = np.linalg.eigvalsh(A_test)  # Symétrique donc valeurs propres réelles

        λ_min = np.min(eigenvalues)
        λ_max = np.max(eigenvalues)
        cond = λ_max / λ_min

        donnees_spectrales.append((n_test, λ_min, λ_max, cond))

        print(f"n={n_test:2d}: λ_min = {λ_min:8.6f}, λ_max = {λ_max:8.6f}, cond(A) = {cond:8.2f}")

        # Graphique des valeurs propres pour chaque dimension
        ax = axes[idx]
        ax.plot(range(1, n_test+1), np.sort(eigenvalues), 'bo-', linewidth=2, markersize=4)
        ax.set_xlabel('Indice de la valeur propre', fontsize=10)
        ax.set_ylabel('Valeur propre', fontsize=10)
        ax.set_title(f'Valeurs propres pour n={n_test}', fontsize=12)
        ax.grid(True, alpha=0.3)
        ax.set_yscale('log' if idx == 5 else 'linear')

    # Graphique conditionnement vs dimension
    ax = axes[5]
    n_vals = [d[0] for d in donnees_spectrales]
    cond_vals = [d[3] for d in donnees_spectrales]
    ax.plot(n_vals, cond_vals, 'ro-', linewidth=2, markersize=8)
    ax.set_xlabel('Dimension n', fontsize=10)
    ax.set_ylabel('Conditionnement κ(A)', fontsize=10)
    ax.set_title('Conditionnement vs Dimension', fontsize=12)
    ax.grid(True, alpha=0.3)
    ax.set_yscale('log')

    plt.suptitle('Question 5: Analyse spectrale de la matrice A', fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('figure2_analyse_spectrale.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Solution théorique pour n=2
    print(f"\nSolution théorique pour n=2:")
    print(f"A = \n{A}")
    print(f"b = {b}")
    u_exact = np.linalg.solve(A, b)
    print(f"u* = A⁻¹b = {u_exact}")

    # ============================================================================
    # QUESTION 6: Méthodes de gradient
    # ============================================================================
    print("\n--- Question 6: Méthodes de gradient ---")

    def gradient_fixe(A, b, u0, rho, eps=1e-6, max_iter=10000):
        """Gradient à pas fixe"""
        u = u0.copy()
        residuals = []
        erreurs = []

        for k in range(max_iter):
            grad = np.dot(A, u) - b
            u_new = u - rho * grad
            residual = np.linalg.norm(u_new - u)
            erreur = np.linalg.norm(grad)

            residuals.append(residual)
            erreurs.append(erreur)

            u = u_new
            if residual < eps:
                break

        return u, k+1, residuals, erreurs

    def gradient_optimal(A, b, u0, eps=1e-6, max_iter=10000):
        """Gradient à pas optimal"""
        u = u0.copy()
        residuals = []
        erreurs = []

        for k in range(max_iter):
            r = b - np.dot(A, u)  # Résidu = -gradient
            Ar = np.dot(A, r)
            alpha = np.dot(r, r) / np.dot(r, Ar)

            u_new = u + alpha * r
            residual = np.linalg.norm(u_new - u)
            erreur = np.linalg.norm(r)

            residuals.append(residual)
            erreurs.append(erreur)

            u = u_new
            if residual < eps:
                break

        return u, k+1, residuals, erreurs

    def gradient_conjugue(A, b, u0, eps=1e-6, max_iter=None):
        """Gradient conjugué"""
        if max_iter is None:
            max_iter = len(b)

        u = u0.copy()
        r = b - np.dot(A, u)
        d = r.copy()
        residuals = [np.linalg.norm(r)]
        erreurs = [np.linalg.norm(r)]

        for k in range(max_iter):
            Ad = np.dot(A, d)
            alpha = np.dot(r, r) / np.dot(d, Ad)
            u_new = u + alpha * d
            r_new = r - alpha * Ad

            residual = np.linalg.norm(u_new - u)
            erreur = np.linalg.norm(r_new)

            residuals.append(residual)
            erreurs.append(erreur)

            u = u_new
            r = r_new

            if residual < eps:
                break

            beta = np.dot(r_new, r_new) / np.dot(r, r)
            d = r_new + beta * d

        return u, k+1, residuals, erreurs

    # ============================================================================
    # VISUALISATION POUR n=2
    # ============================================================================
    print("\nVisualisation pour n=2...")

    # Paramètres pour n=2
    u0_n2 = np.array([-1.0, 2.0])
    rho = 0.02  # Pas fixe adapté
    eps = 1e-6

    # Exécution des méthodes
    u_fixe, iter_fixe, res_fixe, err_fixe = gradient_fixe(A, b, u0_n2, rho, eps)
    u_opt, iter_opt, res_opt, err_opt = gradient_optimal(A, b, u0_n2, eps)
    u_cg, iter_cg, res_cg, err_cg = gradient_conjugue(A, b, u0_n2, eps)

    print(f"Gradient à pas fixe: {iter_fixe} itérations")
    print(f"Gradient à pas optimal: {iter_opt} itérations")
    print(f"Gradient conjugué: {iter_cg} itérations")

    # Figure 3: Trajectoires pour n=2
    fig = plt.figure(figsize=(15, 5))

    # Courbes de niveau avec trajectoires
    ax1 = fig.add_subplot(131)

    # Courbes de niveau
    contour = ax1.contour(U1, U2, Z, levels=30, cmap='viridis', alpha=0.7, linewidths=1)
    ax1.clabel(contour, inline=True, fontsize=8)

    # Champ de gradient
    u1_grad, u2_grad = np.meshgrid(np.linspace(-10, 10, 20), np.linspace(-10, 10, 20))
    grad_u1 = np.zeros_like(u1_grad)
    grad_u2 = np.zeros_like(u2_grad)

    for i in range(u1_grad.shape[0]):
        for j in range(u1_grad.shape[1]):
            u = np.array([u1_grad[i, j], u2_grad[i, j]])
            grad = DJ_n(u, A, b)
            grad_u1[i, j] = grad[0]
            grad_u2[i, j] = grad[1]

    # Normalisation pour une meilleure visualisation
    norm_grad = np.sqrt(grad_u1**2 + grad_u2**2)
    mask = norm_grad > 0
    scale_factor = 0.1

    ax1.quiver(u1_grad[mask], u2_grad[mask],
               -grad_u1[mask]/norm_grad[mask]*scale_factor,
               -grad_u2[mask]/norm_grad[mask]*scale_factor,
               color='red', alpha=0.6, scale=20, width=0.005)

    # Simulation des trajectoires
    def simuler_trajectoire(methode, A, b, u0, n_points=50):
        u = u0.copy()
        traj = [u0.copy()]

        if methode == 'fixe':
            for _ in range(n_points):
                grad = DJ_n(u, A, b)
                u = u - rho * grad
                traj.append(u.copy())
        elif methode == 'optimal':
            for _ in range(n_points):
                r = b - np.dot(A, u)
                Ar = np.dot(A, r)
                alpha = np.dot(r, r) / np.dot(r, Ar)
                u = u + alpha * r
                traj.append(u.copy())
        elif methode == 'conjugue':
            r = b - np.dot(A, u)
            d = r.copy()
            for _ in range(min(n_points, len(b))):
                Ad = np.dot(A, d)
                alpha = np.dot(r, r) / np.dot(d, Ad)
                u = u + alpha * d
                traj.append(u.copy())
                r_new = r - alpha * Ad
                beta = np.dot(r_new, r_new) / np.dot(r, r)
                d = r_new + beta * d
                r = r_new

        return np.array(traj)

    # Trajectoires
    traj_fixe = simuler_trajectoire('fixe', A, b, u0_n2)
    traj_opt = simuler_trajectoire('optimal', A, b, u0_n2)
    traj_cg = simuler_trajectoire('conjugue', A, b, u0_n2)

    ax1.plot(traj_fixe[:, 0], traj_fixe[:, 1], 'r-', marker='o',
             markersize=3, label='Gradient pas fixe', linewidth=2, alpha=0.8)
    ax1.plot(traj_opt[:, 0], traj_opt[:, 1], 'g-', marker='s',
             markersize=3, label='Gradient pas optimal', linewidth=2, alpha=0.8)
    ax1.plot(traj_cg[:, 0], traj_cg[:, 1], 'b-', marker='^',
             markersize=3, label='Gradient conjugué', linewidth=2, alpha=0.8)

    ax1.plot(u0_n2[0], u0_n2[1], 'k*', markersize=15, label='Point initial')
    ax1.plot(u_exact[0], u_exact[1], 'm*', markersize=15, label='Solution exacte')

    ax1.set_xlabel('$u_1$', fontsize=12)
    ax1.set_ylabel('$u_2$', fontsize=12)
    ax1.set_title('Trajectoires des méthodes', fontsize=14)
    ax1.legend(fontsize=10, loc='upper right')
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim([-3, 3])
    ax1.set_ylim([-3, 3])

    # Figure 4: Convergence des résidus
    ax2 = fig.add_subplot(132)

    iterations_fixe = min(100, len(res_fixe))
    iterations_opt = min(100, len(res_opt))
    iterations_cg = min(10, len(res_cg))

    ax2.semilogy(range(iterations_fixe), res_fixe[:iterations_fixe],
                 'r-', label='Gradient pas fixe', linewidth=2)
    ax2.semilogy(range(iterations_opt), res_opt[:iterations_opt],
                 'g-', label='Gradient pas optimal', linewidth=2)
    ax2.semilogy(range(iterations_cg), res_cg[:iterations_cg],
                 'b-', label='Gradient conjugué', linewidth=2)

    ax2.axhline(y=eps, color='k', linestyle='--', alpha=0.5, label=f'Tolérance ({eps})')
    ax2.set_xlabel('Itération', fontsize=12)
    ax2.set_ylabel('Résidu $||u^{(k+1)} - u^{(k)}||$', fontsize=12)
    ax2.set_title('Convergence des résidus', fontsize=14)
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3, which='both')

    # Figure 5: Évolution de l'erreur
    ax3 = fig.add_subplot(133)

    ax3.semilogy(range(len(err_fixe)), err_fixe, 'r-', label='Gradient pas fixe', linewidth=2)
    ax3.semilogy(range(len(err_opt)), err_opt, 'g-', label='Gradient pas optimal', linewidth=2)
    ax3.semilogy(range(len(err_cg)), err_cg, 'b-', label='Gradient conjugué', linewidth=2)

    ax3.axhline(y=eps, color='k', linestyle='--', alpha=0.5, label=f'Tolérance ({eps})')
    ax3.set_xlabel('Itération', fontsize=12)
    ax3.set_ylabel('Erreur $||∇J_n(u^{(k)})||$', fontsize=12)
    ax3.set_title('Évolution de l\'erreur', fontsize=14)
    ax3.legend(fontsize=10)
    ax3.grid(True, alpha=0.3, which='both')

    plt.suptitle('Question 6: Comparaison des méthodes pour n=2', fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('figure3_trajectoires_convergence.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ============================================================================
    # COMPARAISON POUR DIFFÉRENTES DIMENSIONS
    # ============================================================================
    print("\nComparaison pour différentes dimensions...")

    dimensions = [10, 20, 30, 50, 100]
    resultats = []

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    temps_fixe = []
    temps_opt = []
    temps_cg = []
    iter_fixe_list = []
    iter_opt_list = []
    iter_cg_list = []

    for idx, n_dim in enumerate(dimensions):
        print(f"  n={n_dim}...")

        A_dim, h_dim = construire_matrice_A(n_dim)
        b_dim, _ = construire_vecteur_b(n_dim, f_type=1)
        u0_dim = np.zeros(n_dim)

        # Pas adaptatif pour le gradient fixe
        eigvals = np.linalg.eigvalsh(A_dim)
        λ_max = np.max(eigvals)
        ρ_dim = 1.8 / λ_max

        # Gradient à pas fixe
        start = time.time()
        u_fixe_dim, iter_fixe_dim, res_fixe_dim, err_fixe_dim = gradient_fixe(A_dim, b_dim, u0_dim, ρ_dim, eps=1e-6)
        time_fixe = time.time() - start

        # Gradient à pas optimal
        start = time.time()
        u_opt_dim, iter_opt_dim, res_opt_dim, err_opt_dim = gradient_optimal(A_dim, b_dim, u0_dim, eps=1e-6)
        time_opt = time.time() - start

        # Gradient conjugué
        start = time.time()
        u_cg_dim, iter_cg_dim, res_cg_dim, err_cg_dim = gradient_conjugue(A_dim, b_dim, u0_dim, eps=1e-6)
        time_cg = time.time() - start

        resultats.append((n_dim, iter_fixe_dim, iter_opt_dim, iter_cg_dim, time_fixe, time_opt, time_cg))

        # Stockage pour graphiques
        temps_fixe.append(time_fixe)
        temps_opt.append(time_opt)
        temps_cg.append(time_cg)
        iter_fixe_list.append(iter_fixe_dim)
        iter_opt_list.append(iter_opt_dim)
        iter_cg_list.append(iter_cg_dim)

    # Affichage des résultats
    print("\nRésultats de comparaison:")
    print("n\t| Fixe\t| Optimal\t| Conjugué\t| Temps_fixe\t| Temps_opt\t| Temps_cg")
    print("-" * 85)

    for n_dim, iter_f, iter_o, iter_c, t_f, t_o, t_c in resultats:
        print(f"{n_dim}\t| {iter_f:4d}\t| {iter_o:4d}\t\t| {iter_c:4d}\t\t| {t_f:.4f}\t| {t_o:.4f}\t| {t_c:.4f}")

    # Figure 6: Nombre d'itérations vs dimension
    ax1 = axes[0, 0]
    ax1.plot(dimensions, iter_fixe_list, 'ro-', linewidth=2, markersize=8, label='Gradient pas fixe')
    ax1.plot(dimensions, iter_opt_list, 'gs-', linewidth=2, markersize=8, label='Gradient pas optimal')
    ax1.plot(dimensions, iter_cg_list, 'b^-', linewidth=2, markersize=8, label='Gradient conjugué')
    ax1.set_xlabel('Dimension $n$', fontsize=12)
    ax1.set_ylabel('Nombre d\'itérations', fontsize=12)
    ax1.set_title('Nombre d\'itérations vs dimension', fontsize=14)
    ax1.legend(fontsize=10)
    ax1.grid(True, alpha=0.3)
    ax1.set_yscale('log')

    # Figure 7: Temps de calcul vs dimension
    ax2 = axes[0, 1]
    ax2.plot(dimensions, temps_fixe, 'ro-', linewidth=2, markersize=8, label='Gradient pas fixe')
    ax2.plot(dimensions, temps_opt, 'gs-', linewidth=2, markersize=8, label='Gradient pas optimal')
    ax2.plot(dimensions, temps_cg, 'b^-', linewidth=2, markersize=8, label='Gradient conjugué')
    ax2.set_xlabel('Dimension $n$', fontsize=12)
    ax2.set_ylabel('Temps de calcul (s)', fontsize=12)
    ax2.set_title('Temps de calcul vs dimension', fontsize=14)
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)
    ax2.set_yscale('log')

    # Figure 8: Ratio d'efficacité
    ax3 = axes[1, 0]
    ratio_opt_fixe = [t_o/t_f for t_f, t_o in zip(temps_fixe, temps_opt)]
    ratio_cg_fixe = [t_c/t_f for t_f, t_c in zip(temps_fixe, temps_cg)]
    ratio_cg_opt = [t_c/t_o for t_o, t_c in zip(temps_opt, temps_cg)]

    ax3.plot(dimensions, ratio_opt_fixe, 'r--', linewidth=2, label='Optimal/Fixe')
    ax3.plot(dimensions, ratio_cg_fixe, 'g--', linewidth=2, label='Conjugué/Fixe')
    ax3.plot(dimensions, ratio_cg_opt, 'b--', linewidth=2, label='Conjugué/Optimal')
    ax3.axhline(y=1, color='k', linestyle=':', alpha=0.5)
    ax3.set_xlabel('Dimension $n$', fontsize=12)
    ax3.set_ylabel('Ratio de temps', fontsize=12)
    ax3.set_title('Ratio d\'efficacité', fontsize=14)
    ax3.legend(fontsize=10)
    ax3.grid(True, alpha=0.3)

    # Figure 9: Vitesse de convergence théorique
    ax4 = axes[1, 1]

    # Calcul théorique pour le gradient fixe
    taux_convergence = []
    for n_dim in dimensions:
        A_dim, _ = construire_matrice_A(n_dim)
        eigvals = np.linalg.eigvalsh(A_dim)
        λ_min = np.min(eigvals)
        λ_max = np.max(eigvals)

        # Taux de convergence théorique pour gradient fixe avec pas optimal
        κ = λ_max / λ_min
        taux = (κ - 1) / (κ + 1)
        taux_convergence.append(taux)

    ax4.plot(dimensions, taux_convergence, 'mo-', linewidth=2, markersize=8)
    ax4.set_xlabel('Dimension $n$', fontsize=12)
    ax4.set_ylabel('Taux de convergence', fontsize=12)
    ax4.set_title('Taux de convergence théorique', fontsize=14)
    ax4.grid(True, alpha=0.3)

    plt.suptitle('Question 6: Comparaison des performances pour différentes dimensions',
                 fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('figure4_comparaison_dimensions.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("\n✓ Partie I terminée avec succès")
    return True

# ============================================================================
# PARTIE II: OPTIMISATION AVEC CONTRAINTES
# ============================================================================

def partie_II():
    """Exécute toute la Partie II avec toutes les visualisations"""
    print("\n" + "=" * 70)
    print("PARTIE II: OPTIMISATION AVEC CONTRAINTES")
    print("=" * 70)

    # ============================================================================
    # FONCTIONS UTILITAIRES
    # ============================================================================

    def construire_matrice_A(n):
        h = 1.0 / (n + 1)
        main_diag = 2 * np.ones(n)
        off_diag = -1 * np.ones(n-1)
        A = (n+1)**2 * (np.diag(main_diag) + np.diag(off_diag, -1) + np.diag(off_diag, 1))
        return A, h

    def construire_vecteur_b(n, f_type=1):
        h = 1.0 / (n + 1)
        x = np.linspace(0, 1, n+2)[1:-1]

        if f_type == 1:
            f_vals = np.ones(n)
        else:
            f_vals = np.pi**2 * np.sin(np.pi * x)

        b = h * f_vals
        return b, x

    def J_n(u, A, b):
        return 0.5 * np.dot(u, np.dot(A, u)) - np.dot(b, u)

    def DJ_n(u, A, b):
        return np.dot(A, u) - b

    def fonction_obstacle(x):
        return np.maximum(1.5 - 20 * (x - 0.6)**2, 0)

    # ============================================================================
    # QUESTION 1: Résolution directe avec SLSQP
    # ============================================================================
    print("\n--- Question 1: Résolution directe avec SLSQP ---")

    n = 100
    A, h = construire_matrice_A(n)
    x = np.linspace(0, 1, n+2)[1:-1]
    g_vals = fonction_obstacle(x)

    # Cas f(x)=1
    b1, _ = construire_vecteur_b(n, f_type=1)
    u0 = np.zeros(n)
    constraints = ({'type': 'ineq', 'fun': lambda u: u - g_vals})

    print("Résolution pour f(x)=1...")
    start = time.time()
    res1 = minimize(lambda u: J_n(u, A, b1), u0, method='SLSQP',
                    jac=lambda u: DJ_n(u, A, b1), constraints=constraints,
                    tol=1e-8, options={'maxiter': 5000, 'disp': False})
    time1 = time.time() - start

    # Cas f(x)=π²sin(πx)
    b2, _ = construire_vecteur_b(n, f_type=2)
    print("Résolution pour f(x)=π²sin(πx)...")
    start = time.time()
    res2 = minimize(lambda u: J_n(u, A, b2), u0, method='SLSQP',
                    jac=lambda u: DJ_n(u, A, b2), constraints=constraints,
                    tol=1e-8, options={'maxiter': 5000, 'disp': False})
    time2 = time.time() - start

    # Figure 5: Solutions SLSQP
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    # Solution f(x)=1
    ax1 = axes[0, 0]
    ax1.plot(x, res1.x, 'b-', linewidth=2, label='Solution $u_h$')
    ax1.plot(x, g_vals, 'r--', linewidth=2, label='Obstacle $g(x)$')
    ax1.fill_between(x, g_vals, res1.x, where=(res1.x >= g_vals),
                     color='blue', alpha=0.2, label='Zone libre')
    contact_mask1 = np.abs(res1.x - g_vals) < 1e-3
    if np.any(contact_mask1):
        ax1.plot(x[contact_mask1], res1.x[contact_mask1], 'ro',
                 markersize=4, label='Points de contact', alpha=0.7)
    ax1.set_xlabel('$x$', fontsize=12)
    ax1.set_ylabel('$u(x)$', fontsize=12)
    ax1.set_title('Solution pour $f(x)=1$', fontsize=14)
    ax1.legend(fontsize=9)
    ax1.grid(True, alpha=0.3)

    # Solution f(x)=π²sin(πx)
    ax2 = axes[0, 1]
    ax2.plot(x, res2.x, 'b-', linewidth=2, label='Solution $u_h$')
    ax2.plot(x, g_vals, 'r--', linewidth=2, label='Obstacle $g(x)$')
    ax2.fill_between(x, g_vals, res2.x, where=(res2.x >= g_vals),
                     color='blue', alpha=0.2, label='Zone libre')
    contact_mask2 = np.abs(res2.x - g_vals) < 1e-3
    if np.any(contact_mask2):
        ax2.plot(x[contact_mask2], res2.x[contact_mask2], 'ro',
                 markersize=4, label='Points de contact', alpha=0.7)
    ax2.set_xlabel('$x$', fontsize=12)
    ax2.set_ylabel('$u(x)$', fontsize=12)
    ax2.set_title('Solution pour $f(x)=\pi^2\sin(\pi x)$', fontsize=14)
    ax2.legend(fontsize=9)
    ax2.grid(True, alpha=0.3)

    # Multiplicateurs de Lagrange
    def calculer_multiplicateurs(u, A, b):
        λ = -(np.dot(A, u) - b)
        return np.maximum(0, λ)

    λ1 = calculer_multiplicateurs(res1.x, A, b1)
    λ2 = calculer_multiplicateurs(res2.x, A, b2)

    # Produit de complémentarité
    prod1 = λ1 * (res1.x - g_vals)
    prod2 = λ2 * (res2.x - g_vals)

    # Produit de complémentarité f(x)=1
    ax3 = axes[1, 0]
    ax3.plot(x, prod1, 'b-', linewidth=1.5)
    ax3.fill_between(x, 0, prod1, where=(prod1 > 0), color='blue', alpha=0.3)
    ax3.axhline(y=0, color='r', linestyle='--', alpha=0.5, linewidth=1)
    ax3.set_xlabel('$x$', fontsize=12)
    ax3.set_ylabel('$(Au-b+\lambda)(u-g)$', fontsize=12)
    ax3.set_title('Produit de complémentarité pour $f(x)=1$', fontsize=14)
    ax3.set_ylim([-1e-13, max(prod1)*1.1])
    ax3.grid(True, alpha=0.3)

    # Produit de complémentarité f(x)=π²sin(πx)
    ax4 = axes[1, 1]
    ax4.plot(x, prod2, 'b-', linewidth=1.5)
    ax4.fill_between(x, 0, prod2, where=(prod2 > 0), color='blue', alpha=0.3)
    ax4.axhline(y=0, color='r', linestyle='--', alpha=0.5, linewidth=1)
    ax4.set_xlabel('$x$', fontsize=12)
    ax4.set_ylabel('$(Au-b+\lambda)(u-g)$', fontsize=12)
    ax4.set_title('Produit de complémentarité pour $f(x)=\pi^2\sin(\pi x)$', fontsize=14)
    ax4.set_ylim([-1e-13, max(prod2)*1.1])
    ax4.grid(True, alpha=0.3)

    plt.suptitle('Question 1: Résolution directe du problème d\'obstacle',
                 fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('figure5_solutions_obstacle.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Vérification KKT
    print(f"\nVérification des conditions KKT:")
    print(f"Cas f(x)=1: max|Au-b+λ| = {np.max(np.abs(np.dot(A, res1.x) - b1 + λ1)):.2e}")
    print(f"            min(u-g) = {np.min(res1.x - g_vals):.2e}")
    print(f"            max|λ(u-g)| = {np.max(np.abs(prod1)):.2e}")

    print(f"\nCas f(x)=π²sin(πx): max|Au-b+λ| = {np.max(np.abs(np.dot(A, res2.x) - b2 + λ2)):.2e}")
    print(f"                    min(u-g) = {np.min(res2.x - g_vals):.2e}")
    print(f"                    max|λ(u-g)| = {np.max(np.abs(prod2)):.2e}")

    # ============================================================================
    # QUESTION 2: Méthode du gradient projeté
    # ============================================================================
    print("\n--- Question 2: Méthode du gradient projeté ---")

    def projection_K(u, g):
        return np.maximum(g, u)

    def gradient_projete(A, b, g, u0, rho, eps=1e-6, max_iter=5000):
        u = u0.copy()
        residuals = []
        violations = []

        for k in range(max_iter):
            grad = np.dot(A, u) - b
            u_temp = u - rho * grad
            u_new = projection_K(u_temp, g)

            residual = np.linalg.norm(u_new - u)
            violation = np.min(u_new - g)

            residuals.append(residual)
            violations.append(violation)

            u = u_new
            if residual < eps:
                break

        return u, k+1, residuals, violations

    # Analyse pour différentes dimensions
    dimensions_GP = [10, 20, 50, 100]
    results_GP = []

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()

    for idx, n_dim in enumerate(dimensions_GP[:5]):
        A_dim, _ = construire_matrice_A(n_dim)
        b_dim, _ = construire_vecteur_b(n_dim, f_type=1)
        x_dim = np.linspace(0, 1, n_dim+2)[1:-1]
        g_dim = fonction_obstacle(x_dim)
        u0_dim = np.zeros(n_dim)

        # Calcul du pas optimal
        eigvals = np.linalg.eigvalsh(A_dim)
        λ_min = np.min(eigvals)
        λ_max = np.max(eigvals)
        ρ_opt = 2.0 / (λ_min + λ_max)
        ρ_safe = 1.8 / λ_max

        print(f"n={n_dim}: λ_min={λ_min:.4f}, λ_max={λ_max:.4f}, ρ_opt={ρ_opt:.6f}, ρ_safe={ρ_safe:.6f}")

        u_gp, iter_gp, res_gp, viol_gp = gradient_projete(A_dim, b_dim, g_dim, u0_dim,
                                                          rho=ρ_safe, eps=1e-6, max_iter=5000)

        results_GP.append((n_dim, ρ_safe, iter_gp, np.min(viol_gp)))

        # Visualisation
        ax = axes[idx]
        ax.plot(x_dim, u_gp, 'b-', linewidth=2, label='Solution GP')
        ax.plot(x_dim, g_dim, 'r--', linewidth=1.5, label='Obstacle')
        ax.fill_between(x_dim, g_dim, u_gp, where=(u_gp >= g_dim),
                        color='blue', alpha=0.2)
        ax.set_xlabel('$x$', fontsize=10)
        ax.set_ylabel('$u(x)$', fontsize=10)
        ax.set_title(f'n={n_dim}, ρ={ρ_safe:.4f}\n{iter_gp} itérations', fontsize=11)
        ax.legend(fontsize=8)
        ax.grid(True, alpha=0.3)

    # Analyse de convergence
    ax = axes[5]
    for i, (n_dim, ρ, iter_gp, viol) in enumerate(results_GP[:3]):
        A_dim, _ = construire_matrice_A(n_dim)
        b_dim, _ = construire_vecteur_b(n_dim, f_type=1)
        x_dim = np.linspace(0, 1, n_dim+2)[1:-1]
        g_dim = fonction_obstacle(x_dim)
        u0_dim = np.zeros(n_dim)

        u_gp, iter_gp, res_gp, viol_gp = gradient_projete(A_dim, b_dim, g_dim, u0_dim,
                                                          rho=ρ, eps=1e-6, max_iter=2000)

        n_plot = min(500, len(res_gp))
        ax.semilogy(range(n_plot), res_gp[:n_plot], linewidth=2,
                   label=f'n={n_dim}, ρ={ρ:.4f}')

    ax.set_xlabel('Itération', fontsize=12)
    ax.set_ylabel('Résidu $||u^{(k+1)} - u^{(k)}||$', fontsize=12)
    ax.set_title('Convergence du gradient projeté', fontsize=14)
    ax.legend(fontsize=10)
    ax.grid(True, alpha=0.3, which='both')

    plt.suptitle('Question 2: Méthode du gradient projeté', fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('figure6_gradient_projete.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ============================================================================
    # QUESTION 3: Méthode de pénalisation
    # ============================================================================
    print("\n--- Question 3: Méthode de pénalisation ---")

    def penalisation(A, b, g, u0, eta, rho=None, eps=1e-6, max_iter=5000):
        n = len(b)
        u = u0.copy()
        residuals = []
        violations = []

        if rho is None:
            eigvals = np.linalg.eigvalsh(A)
            λ_max = np.max(eigvals)
            rho_base = 1.8 / λ_max
            rho = min(rho_base, eta/100)

        for k in range(max_iter):
            penal = np.maximum(g - u, 0)
            grad = np.dot(A, u) - b - (2.0/eta) * penal

            u_new = u - rho * grad

            residual = np.linalg.norm(u_new - u)
            violation = np.min(u_new - g)

            residuals.append(residual)
            violations.append(violation)

            if residual < eps and abs(violation) < 1e-4:
                break

            u = u_new

        return u, k+1, residuals, violations

    # Test avec différents η
    eta_vals = [1e6, 1e4, 1e2, 1, 0.1, 0.01]
    results_pen = []

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()

    for idx, eta in enumerate(eta_vals[:6]):
        u_pen, iter_pen, res_pen, viol_pen = penalisation(A, b1, g_vals, u0, eta,
                                                         eps=1e-6, max_iter=5000)

        violation_finale = np.min(u_pen - g_vals)
        results_pen.append((eta, iter_pen, violation_finale, J_n(u_pen, A, b1)))

        print(f"η={eta:.1e}: {iter_pen} itérations, min(u-g)={violation_finale:.2e}, J={J_n(u_pen, A, b1):.6f}")

        # Visualisation
        ax = axes[idx]
        ax.plot(x, u_pen, 'b-', linewidth=2, label=f'η={eta:.1e}')
        ax.plot(x, g_vals, 'r--', linewidth=1.5, label='Obstacle')

        # Zones de violation
        viol_mask = (u_pen - g_vals) < 0
        if np.any(viol_mask):
            ax.plot(x[viol_mask], u_pen[viol_mask], 'rx',
                   markersize=4, label='Violation')

        ax.set_xlabel('$x$', fontsize=10)
        ax.set_ylabel('$u(x)$', fontsize=10)
        ax.set_title(f'η={eta:.1e}, {iter_pen} itérations\nmin(u-g)={violation_finale:.2e}',
                    fontsize=11)
        ax.legend(fontsize=8)
        ax.grid(True, alpha=0.3)

    plt.suptitle('Question 3: Méthode de pénalisation - Influence de η',
                 fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('figure7_penalisation.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ============================================================================
    # QUESTION 4: Méthode d'Uzawa
    # ============================================================================
    print("\n--- Question 4: Méthode d'Uzawa ---")

    def uzawa(A, b, g, u0, lambda0, rho, eps=1e-5, max_iter=100):
        n = len(b)
        u = u0.copy()
        lam = lambda0.copy()
        residuals = []

        for k in range(max_iter):
            # Résolution du sous-problème primal
            b_mod = b - lam
            u_new = np.linalg.solve(A, b_mod)

            # Mise à jour des multiplicateurs
            lam_new = np.maximum(0, lam + rho * (g - u_new))

            residual = np.linalg.norm(u_new - u)
            residuals.append(residual)

            u = u_new
            lam = lam_new

            if residual < eps:
                break

        return u, lam, k+1, residuals

    # Test avec différents ρ
    rho_vals = [0.01, 0.05, 0.1, 0.5, 1.0, 2.0]
    results_uz = []
    lambda0 = np.ones(n) * 0.1

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()

    for idx, rho in enumerate(rho_vals[:6]):
        u_uz, lam_uz, iter_uz, res_uz = uzawa(A, b1, g_vals, u0, lambda0, rho,
                                             eps=1e-5, max_iter=100)

        violation = np.min(u_uz - g_vals)
        active_constraints = np.sum(lam_uz > 1e-3)
        results_uz.append((rho, iter_uz, violation, active_constraints))

        print(f"ρ={rho:.2f}: {iter_uz} itérations, {active_constraints} contraintes actives, min(u-g)={violation:.2e}")

        # Visualisation
        ax = axes[idx]
        ax.plot(x, u_uz, 'b-', linewidth=2, label='Solution')
        ax.plot(x, g_vals, 'r--', linewidth=1.5, label='Obstacle')

        # Points de contact
        contact_mask = lam_uz > 1e-3
        if np.any(contact_mask):
            ax.plot(x[contact_mask], u_uz[contact_mask], 'go',
                   markersize=4, label='Contact (λ>0)', alpha=0.7)

        ax.set_xlabel('$x$', fontsize=10)
        ax.set_ylabel('$u(x)$', fontsize=10)
        ax.set_title(f'ρ={rho:.2f}, {iter_uz} itérations\n{active_constraints} contacts',
                    fontsize=11)
        ax.legend(fontsize=8)
        ax.grid(True, alpha=0.3)

    plt.suptitle('Question 4: Méthode d\'Uzawa - Influence de ρ',
                 fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('figure8_uzawa.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ============================================================================
    # COMPARAISON FINALE DES MÉTHODES
    # ============================================================================
    print("\n--- Comparaison finale des méthodes ---")

    # Solution SLSQP (référence)
    u_ref = res1.x

    # Gradient projeté (avec ρ optimal)
    eigvals = np.linalg.eigvalsh(A)
    λ_min = np.min(eigvals)
    λ_max = np.max(eigvals)
    ρ_gp = 1.8 / λ_max
    u_gp, iter_gp, _, _ = gradient_projete(A, b1, g_vals, u0, ρ_gp, eps=1e-6)

    # Pénalisation (avec η=1)
    η_pen = 1.0
    u_pen, iter_pen, _, _ = penalisation(A, b1, g_vals, u0, η_pen, eps=1e-6)

    # Uzawa (avec ρ=0.1)
    ρ_uz = 0.1
    u_uz, _, iter_uz, _ = uzawa(A, b1, g_vals, u0, lambda0, ρ_uz, eps=1e-5)

    # Figure de comparaison
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    # Solutions
    ax1 = axes[0, 0]
    ax1.plot(x, u_ref, 'k-', linewidth=3, label='SLSQP (référence)', alpha=0.8)
    ax1.plot(x, u_gp, 'r--', linewidth=2, label=f'GP (ρ={ρ_gp:.4f})')
    ax1.plot(x, u_pen, 'g-.', linewidth=2, label=f'Pénalisation (η={η_pen})')
    ax1.plot(x, u_uz, 'b:', linewidth=2, label=f'Uzawa (ρ={ρ_uz})')
    ax1.plot(x, g_vals, 'm-', linewidth=1.5, alpha=0.5, label='Obstacle')
    ax1.set_xlabel('$x$', fontsize=12)
    ax1.set_ylabel('$u(x)$', fontsize=12)
    ax1.set_title('Comparaison des solutions', fontsize=14)
    ax1.legend(fontsize=10, loc='upper right')
    ax1.grid(True, alpha=0.3)

    # Erreurs relatives
    ax2 = axes[0, 1]
    erreur_gp = np.linalg.norm(u_gp - u_ref) / np.linalg.norm(u_ref)
    erreur_pen = np.linalg.norm(u_pen - u_ref) / np.linalg.norm(u_ref)
    erreur_uz = np.linalg.norm(u_uz - u_ref) / np.linalg.norm(u_ref)

    methodes = ['GP', 'Pénalisation', 'Uzawa']
    erreurs = [erreur_gp, erreur_pen, erreur_uz]

    bars = ax2.bar(methodes, erreurs, color=['red', 'green', 'blue'], alpha=0.7)
    ax2.set_ylabel('Erreur relative', fontsize=12)
    ax2.set_title('Erreurs relatives par rapport à SLSQP', fontsize=14)
    ax2.grid(True, alpha=0.3, axis='y')

    # Ajouter les valeurs sur les barres
    for bar, err in zip(bars, erreurs):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.0001,
                f'{err:.2e}', ha='center', va='bottom', fontsize=10)

    # Violations des contraintes
    ax3 = axes[1, 0]
    violation_gp = np.max(np.abs(np.minimum(u_gp - g_vals, 0)))
    violation_pen = np.max(np.abs(np.minimum(u_pen - g_vals, 0)))
    violation_uz = np.max(np.abs(np.minimum(u_uz - g_vals, 0)))

    violations = [violation_gp, violation_pen, violation_uz]

    bars = ax3.bar(methodes, violations, color=['red', 'green', 'blue'], alpha=0.7)
    ax3.set_ylabel('Violation max $|min(u-g)|$', fontsize=12)
    ax3.set_title('Violations des contraintes', fontsize=14)
    ax3.grid(True, alpha=0.3, axis='y')
    ax3.set_yscale('log')

    # Ajouter les valeurs sur les barres
    for bar, viol in zip(bars, violations):
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2., height * 1.2,
                f'{viol:.2e}', ha='center', va='bottom', fontsize=10)

    # Performances (itérations)
    ax4 = axes[1, 1]
    iterations = [iter_gp, iter_pen, iter_uz]

    bars = ax4.bar(methodes, iterations, color=['red', 'green', 'blue'], alpha=0.7)
    ax4.set_ylabel('Nombre d\'itérations', fontsize=12)
    ax4.set_title('Performances (itérations)', fontsize=14)
    ax4.grid(True, alpha=0.3, axis='y')

    # Ajouter les valeurs sur les barres
    for bar, iter_val in zip(bars, iterations):
        height = bar.get_height()
        ax4.text(bar.get_x() + bar.get_width()/2., height + 5,
                f'{iter_val}', ha='center', va='bottom', fontsize=10)

    plt.suptitle('Comparaison finale des méthodes pour $n=100$',
                 fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig('figure9_comparaison_methodes.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Tableau récapitulatif
    print("\n" + "="*80)
    print("RÉCAPITULATIF DES PERFORMANCES (n=100)")
    print("="*80)
    print(f"{'Méthode':<20} | {'Itérations':>12} | {'Erreur relative':>15} | {'Violation max':>13}")
    print("-" * 80)
    print(f"{'SLSQP (réf)':<20} | {res1.nit:>12} | {'0.00e+00':>15} | {np.max(np.abs(np.minimum(u_ref - g_vals, 0))):13.2e}")
    print(f"{'Gradient projeté':<20} | {iter_gp:>12} | {erreur_gp:>15.2e} | {violation_gp:13.2e}")
    print(f"{'Pénalisation':<20} | {iter_pen:>12} | {erreur_pen:>15.2e} | {violation_pen:13.2e}")
    print(f"{'Uzawa':<20} | {iter_uz:>12} | {erreur_uz:>15.2e} | {violation_uz:13.2e}")
    print("="*80)

    print("\n✓ Partie II terminée avec succès")
    return True

# ============================================================================
# PARTIE III: OPTIMISATION DES STRUCTURES DE TENSÉGRITÉ
# ============================================================================

def partie_III():
    """Exécute toute la Partie III avec toutes les visualisations"""
    print("\n" + "=" * 70)
    print("PARTIE III: OPTIMISATION DES STRUCTURES DE TENSÉGRITÉ")
    print("=" * 70)

    # ============================================================================
    # CLASSE STRUCTURE DE TENSÉGRITÉ
    # ============================================================================

    class StructureTensegrity:
        """Classe pour modéliser une structure de tenségrité"""

        def __init__(self):
            self.nodes = {}
            self.bars = []
            self.cables = []
            self.fixed_nodes = {}
            self.masses = {}
            self.g = 9.81
            self.rho = 1.0

        def add_node(self, node_id, position):
            self.nodes[node_id] = np.array(position, dtype=float)

        def add_fixed_node(self, node_id, position):
            self.fixed_nodes[node_id] = np.array(position, dtype=float)
            self.nodes[node_id] = np.array(position, dtype=float)

        def add_bar(self, id1, id2, length, C=1.0):
            self.bars.append((id1, id2, length, C))

        def add_cable(self, id1, id2, length, K=1.0):
            self.cables.append((id1, id2, length, K))

        def add_mass(self, node_id, mass):
            self.masses[node_id] = mass

        def get_free_nodes_ids(self):
            fixed_ids = set(self.fixed_nodes.keys())
            all_ids = set(self.nodes.keys())
            return sorted(list(all_ids - fixed_ids))

        def set_node_positions(self, X):
            free_ids = self.get_free_nodes_ids()
            n_free = len(free_ids)
            for i, node_id in enumerate(free_ids):
                self.nodes[node_id] = X[3*i:3*(i+1)]

        def get_node_positions_vector(self):
            free_ids = self.get_free_nodes_ids()
            X = []
            for node_id in free_ids:
                X.extend(self.nodes[node_id])
            return np.array(X)

    # ============================================================================
    # FONCTION DE VISUALISATION
    # ============================================================================

    def visualiser_structure(structure, title, filename, elev=25, azim=45):
        """Visualise une structure de tenségrité en 3D"""
        fig = plt.figure(figsize=(12, 10))
        ax = fig.add_subplot(111, projection='3d')

        # Positions
        positions = structure.nodes

        # Nœuds fixes et libres
        xs_fixed, ys_fixed, zs_fixed = [], [], []
        xs_free, ys_free, zs_free = [], [], []

        for node_id, pos in positions.items():
            if node_id in structure.fixed_nodes:
                xs_fixed.append(pos[0])
                ys_fixed.append(pos[1])
                zs_fixed.append(pos[2])
            else:
                xs_free.append(pos[0])
                ys_free.append(pos[1])
                zs_free.append(pos[2])

        # Tracer les nœuds
        if xs_fixed:
            ax.scatter(xs_fixed, ys_fixed, zs_fixed, c='red', s=100,
                      alpha=0.9, label='Nœuds fixes', depthshade=True)
        if xs_free:
            ax.scatter(xs_free, ys_free, zs_free, c='blue', s=80,
                      alpha=0.9, label='Nœuds libres', depthshade=True)

        # Tracer les barres (lignes continues épaisses)
        for (id1, id2, l0, C) in structure.bars:
            if id1 in positions and id2 in positions:
                p1 = positions[id1]
                p2 = positions[id2]
                ax.plot([p1[0], p2[0]], [p1[1], p2[1]], [p1[2], p2[2]],
                       'k-', linewidth=3, alpha=0.8, label='Barre' if id1==structure.bars[0][0] else "")

        # Tracer les câbles (lignes pointillées)
        for (id1, id2, l0, K) in structure.cables:
            if id1 in positions and id2 in positions:
                p1 = positions[id1]
                p2 = positions[id2]
                ax.plot([p1[0], p2[0]], [p1[1], p2[1]], [p1[2], p2[2]],
                       'b--', linewidth=1.5, alpha=0.6, label='Câble' if id1==structure.cables[0][0] else "")

        # Configuration du graphique
        ax.set_xlabel('X', fontsize=12, labelpad=10)
        ax.set_ylabel('Y', fontsize=12, labelpad=10)
        ax.set_zlabel('Z', fontsize=12, labelpad=10)
        ax.set_title(title, fontsize=16, pad=20)

        # Ajuster les limites
        all_coords = np.array(list(positions.values()))
        if len(all_coords) > 0:
            max_range = np.max([all_coords[:, i].max() - all_coords[:, i].min() for i in range(3)]) * 0.6
            mid_x = (all_coords[:, 0].max() + all_coords[:, 0].min()) * 0.5
            mid_y = (all_coords[:, 1].max() + all_coords[:, 1].min()) * 0.5
            mid_z = (all_coords[:, 2].max() + all_coords[:, 2].min()) * 0.5

            ax.set_xlim(mid_x - max_range, mid_x + max_range)
            ax.set_ylim(mid_y - max_range, mid_y + max_range)
            ax.set_zlim(mid_z - max_range, mid_z + max_range)

        # Légende
        handles, labels = ax.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        ax.legend(by_label.values(), by_label.keys(), loc='upper left', fontsize=10)

        # Vue
        ax.view_init(elev=elev, azim=azim)

        # Quadrillage
        ax.xaxis._axinfo["grid"]['color'] = (0.5, 0.5, 0.5, 0.1)
        ax.yaxis._axinfo["grid"]['color'] = (0.5, 0.5, 0.5, 0.1)
        ax.zaxis._axinfo["grid"]['color'] = (0.5, 0.5, 0.5, 0.1)

        plt.tight_layout()
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        plt.show()

    # ============================================================================
    # QUESTION 2: RÉSEAUX DE CÂBLES
    # ============================================================================
    print("\n--- Question 2: Réseaux de câbles ---")

    def creer_reseau_cables():
        """Crée un réseau de câbles de test"""
        structure = StructureTensegrity()

        # Nœuds fixes
        structure.add_fixed_node(1, [5, 5, 0])
        structure.add_fixed_node(2, [-5, 5, 0])
        structure.add_fixed_node(3, [-5, -5, 0])
        structure.add_fixed_node(4, [5, -5, 0])

        # Nœuds libres (initialisation au centre)
        structure.add_node(5, [0, 0, 0])
        structure.add_node(6, [0, 0, 0])
        structure.add_node(7, [0, 0, 0])
        structure.add_node(8, [0, 0, 0])

        # Câbles (toutes les connexions)
        K = 3.0
        l0 = 3.0
        edges = [(1,5), (1,6), (1,7), (1,8),
                 (2,5), (2,6), (2,7), (2,8),
                 (3,5), (3,6), (3,7), (3,8),
                 (4,5), (4,6), (4,7), (4,8)]

        for id1, id2 in edges:
            structure.add_cable(id1, id2, l0, K)

        # Masses
        mi = (1/6) / structure.g
        for i in range(5, 9):
            structure.add_mass(i, mi)

        return structure

    def energie_reseau_cables(X, structure):
        """Énergie totale pour un réseau de câbles"""
        structure.set_node_positions(X)
        E_total = 0.0

        # Énergie élastique des câbles
        for (id1, id2, l0, K) in structure.cables:
            pos1 = structure.nodes[id1]
            pos2 = structure.nodes[id2]
            dist = np.linalg.norm(pos1 - pos2)

            if dist > l0:
                E_total += (K / (2 * l0**2)) * (dist - l0)**2

        # Énergie gravitationnelle
        for node_id, mass in structure.masses.items():
            if node_id in structure.nodes:
                z = structure.nodes[node_id][2]
                E_total += mass * structure.g * z

        return E_total

    def gradient_reseau_cables(X, structure, h=1e-6):
        """Gradient par différences finies"""
        n = len(X)
        grad = np.zeros(n)

        for i in range(n):
            X_plus = X.copy()
            X_plus[i] += h
            E_plus = energie_reseau_cables(X_plus, structure)

            X_minus = X.copy()
            X_minus[i] -= h
            E_minus = energie_reseau_cables(X_minus, structure)

            grad[i] = (E_plus - E_minus) / (2 * h)

        return grad

    # Optimisation
    print("Optimisation du réseau de câbles...")

    structure_cables = creer_reseau_cables()
    X0_cables = structure_cables.get_node_positions_vector()

    start = time.time()
    result_cables = minimize(
        fun=lambda X: energie_reseau_cables(X, structure_cables),
        x0=X0_cables,
        method='BFGS',
        jac=lambda X: gradient_reseau_cables(X, structure_cables),
        options={'gtol': 1e-8, 'maxiter': 1000, 'disp': False}
    )
    time_cables = time.time() - start

    structure_cables.set_node_positions(result_cables.x)

    print(f"Terminé en {result_cables.nit} itérations, temps: {time_cables:.3f}s")
    print(f"Énergie finale: {result_cables.fun:.6f}")

    # Solution analytique de référence
    solution_ref_cables = {
        5: np.array([2, 2, -1.5]),
        6: np.array([-2, 2, -1.5]),
        7: np.array([-2, -2, -1.5]),
        8: np.array([2, -2, -1.5])
    }

    print("\nPositions optimales vs solution analytique:")
    erreurs_cables = []
    for node_id in range(5, 9):
        pos_opt = structure_cables.nodes[node_id]
        pos_ref = solution_ref_cables[node_id]
        erreur = np.linalg.norm(pos_opt - pos_ref)
        erreurs_cables.append(erreur)

        print(f"  x({node_id}): [{pos_opt[0]:8.4f}, {pos_opt[1]:8.4f}, {pos_opt[2]:8.4f}]")
        print(f"     Référence: [{pos_ref[0]:8.4f}, {pos_ref[1]:8.4f}, {pos_ref[2]:8.4f}]")
        print(f"     Erreur: {erreur:.6f}")

    print(f"\nErreur moyenne: {np.mean(erreurs_cables):.6f}")

    # Vérification du gradient
    grad_norm = np.linalg.norm(gradient_reseau_cables(result_cables.x, structure_cables))
    print(f"Norme du gradient à l'optimum: {grad_norm:.2e}")

    # Visualisation
    visualiser_structure(structure_cables,
                        "Réseau de câbles - Solution optimale",
                        "figure10_reseau_cables.png", elev=30, azim=45)

    # ============================================================================
    # QUESTION 3: DÔMES DE TENSÉGRITÉ
    # ============================================================================
    print("\n--- Question 3: Dômes de tenségrité ---")

    def creer_dome_tensegrity():
        """Crée un dôme de tenségrité de test"""
        structure = StructureTensegrity()
        structure.rho = 0.0  # gρ = 0 selon l'énoncé

        # Nœuds fixes
        structure.add_fixed_node(1, [1, 1, 0])
        structure.add_fixed_node(2, [-1, 1, 0])
        structure.add_fixed_node(3, [-1, -1, 0])
        structure.add_fixed_node(4, [1, -1, 0])

        # Nœuds libres (initialisation haute)
        structure.add_node(5, [-0.5, 0, 8])
        structure.add_node(6, [0, -0.5, 8])
        structure.add_node(7, [0.5, 0, 8])
        structure.add_node(8, [0, 0.5, 8])

        # Barres
        c = 1.0
        structure.add_bar(1, 5, 10, c)
        structure.add_bar(2, 6, 10, c)
        structure.add_bar(3, 7, 10, c)
        structure.add_bar(4, 8, 10, c)

        # Câbles
        k = 0.1
        structure.add_cable(1, 8, 8, k)
        structure.add_cable(2, 5, 8, k)
        structure.add_cable(3, 6, 8, k)
        structure.add_cable(4, 7, 8, k)

        # Câbles entre nœuds libres
        structure.add_cable(5, 6, 1, k)
        structure.add_cable(6, 7, 1, k)
        structure.add_cable(7, 8, 1, k)
        structure.add_cable(5, 8, 1, k)

        return structure

    def energie_dome_tensegrity(X, structure):
        """Énergie totale pour un dôme de tenségrité"""
        structure.set_node_positions(X)
        E_total = 0.0

        # Énergie des barres
        for (id1, id2, l0, C) in structure.bars:
            pos1 = structure.nodes[id1]
            pos2 = structure.nodes[id2]
            dist = np.linalg.norm(pos1 - pos2)
            E_total += (C / (2 * l0**2)) * (dist - l0)**2

        # Énergie des câbles
        for (id1, id2, l0, K) in structure.cables:
            pos1 = structure.nodes[id1]
            pos2 = structure.nodes[id2]
            dist = np.linalg.norm(pos1 - pos2)

            if dist > l0:
                E_total += (K / (2 * l0**2)) * (dist - l0)**2

        return E_total

    def gradient_dome_tensegrity(X, structure, h=1e-6):
        """Gradient par différences finies"""
        n = len(X)
        grad = np.zeros(n)

        for i in range(n):
            X_plus = X.copy()
            X_plus[i] += h
            E_plus = energie_dome_tensegrity(X_plus, structure)

            X_minus = X.copy()
            X_minus[i] -= h
            E_minus = energie_dome_tensegrity(X_minus, structure)

            grad[i] = (E_plus - E_minus) / (2 * h)

        return grad

    # Optimisation
    print("Optimisation du dôme de tenségrité...")

    structure_dome = creer_dome_tensegrity()
    X0_dome = structure_dome.get_node_positions_vector()

    start = time.time()
    result_dome = minimize(
        fun=lambda X: energie_dome_tensegrity(X, structure_dome),
        x0=X0_dome,
        method='BFGS',
        jac=lambda X: gradient_dome_tensegrity(X, structure_dome),
        options={'gtol': 1e-6, 'maxiter': 2000, 'disp': False}
    )
    time_dome = time.time() - start

    structure_dome.set_node_positions(result_dome.x)

    print(f"Terminé en {result_dome.nit} itérations, temps: {time_dome:.3f}s")
    print(f"Énergie finale: {result_dome.fun:.6f}")

    # Solution analytique de référence
    s_ref = 0.70970
    t_ref = 9.54287
    solution_ref_dome = {
        5: np.array([-s_ref, 0, t_ref]),
        6: np.array([0, -s_ref, t_ref]),
        7: np.array([s_ref, 0, t_ref]),
        8: np.array([0, s_ref, t_ref])
    }

    print("\nPositions optimales vs solution analytique:")
    erreurs_dome = []
    hauteurs = []

    for node_id in range(5, 9):
        pos_opt = structure_dome.nodes[node_id]
        pos_ref = solution_ref_dome[node_id]
        erreur = np.linalg.norm(pos_opt - pos_ref)
        erreurs_dome.append(erreur)
        hauteurs.append(pos_opt[2])

        print(f"  x({node_id}): [{pos_opt[0]:8.5f}, {pos_opt[1]:8.5f}, {pos_opt[2]:8.5f}]")
        print(f"     Référence: [{pos_ref[0]:8.5f}, {pos_ref[1]:8.5f}, {pos_ref[2]:8.5f}]")
        print(f"     Erreur: {erreur:.6f}")

    print(f"\nErreur moyenne: {np.mean(erreurs_dome):.6f}")
    print(f"Hauteur moyenne: {np.mean(hauteurs):.5f} (référence: {t_ref:.5f})")

    # Vérification du gradient
    grad_norm = np.linalg.norm(gradient_dome_tensegrity(result_dome.x, structure_dome))
    print(f"Norme du gradient à l'optimum: {grad_norm:.2e}")

    # Visualisation
    visualiser_structure(structure_dome,
                        "Dôme de tenségrité - Solution optimale",
                        "figure11_dome_tensegrity.png", elev=25, azim=45)

    # ============================================================================
    # COMPARAISON ET ANALYSE
    # ============================================================================
    print("\n--- Analyse comparative ---")

    # Figure de comparaison
    fig = plt.figure(figsize=(16, 8))

    # Réseau de câbles
    ax1 = fig.add_subplot(121, projection='3d')
    positions_cables = structure_cables.nodes

    # Nœuds
    xs_fixed_c, ys_fixed_c, zs_fixed_c = [], [], []
    xs_free_c, ys_free_c, zs_free_c = [], [], []

    for node_id, pos in positions_cables.items():
        if node_id in structure_cables.fixed_nodes:
            xs_fixed_c.append(pos[0])
            ys_fixed_c.append(pos[1])
            zs_fixed_c.append(pos[2])
        else:
            xs_free_c.append(pos[0])
            ys_free_c.append(pos[1])
            zs_free_c.append(pos[2])

    ax1.scatter(xs_fixed_c, ys_fixed_c, zs_fixed_c, c='red', s=100, alpha=0.9)
    ax1.scatter(xs_free_c, ys_free_c, zs_free_c, c='blue', s=80, alpha=0.9)

    # Câbles
    for (id1, id2, l0, K) in structure_cables.cables:
        p1, p2 = positions_cables[id1], positions_cables[id2]
        ax1.plot([p1[0], p2[0]], [p1[1], p2[1]], [p1[2], p2[2]],
                'b--', linewidth=1, alpha=0.6)

    ax1.set_title("Réseau de câbles", fontsize=14)
    ax1.set_xlabel('X'); ax1.set_ylabel('Y'); ax1.set_zlabel('Z')
    ax1.view_init(elev=20, azim=45)

    # Dôme de tenségrité
    ax2 = fig.add_subplot(122, projection='3d')
    positions_dome = structure_dome.nodes

    xs_fixed_d, ys_fixed_d, zs_fixed_d = [], [], []
    xs_free_d, ys_free_d, zs_free_d = [], [], []

    for node_id, pos in positions_dome.items():
        if node_id in structure_dome.fixed_nodes:
            xs_fixed_d.append(pos[0])
            ys_fixed_d.append(pos[1])
            zs_fixed_d.append(pos[2])
        else:
            xs_free_d.append(pos[0])
            ys_free_d.append(pos[1])
            zs_free_d.append(pos[2])

    ax2.scatter(xs_fixed_d, ys_fixed_d, zs_fixed_d, c='red', s=100, alpha=0.9)
    ax2.scatter(xs_free_d, ys_free_d, zs_free_d, c='blue', s=80, alpha=0.9)

    # Barres
    for (id1, id2, l0, C) in structure_dome.bars:
        p1, p2 = positions_dome[id1], positions_dome[id2]
        ax2.plot([p1[0], p2[0]], [p1[1], p2[1]], [p1[2], p2[2]],
                'k-', linewidth=3, alpha=0.8)

    # Câbles
    for (id1, id2, l0, K) in structure_dome.cables:
        p1, p2 = positions_dome[id1], positions_dome[id2]
        ax2.plot([p1[0], p2[0]], [p1[1], p2[1]], [p1[2], p2[2]],
                'b--', linewidth=1.5, alpha=0.6)

    ax2.set_title("Dôme de tenségrité", fontsize=14)
    ax2.set_xlabel('X'); ax2.set_ylabel('Y'); ax2.set_zlabel('Z')
    ax2.view_init(elev=20, azim=45)

    plt.suptitle("Comparaison des structures de tenségrité", fontsize=16, y=0.98)
    plt.tight_layout()
    plt.savefig("figure12_comparaison_structures.png", dpi=300, bbox_inches='tight')
    plt.show()

    # Tableau récapitulatif
    print("\n" + "="*80)
    print("RÉCAPITULATIF DES RÉSULTATS - PARTIE III")
    print("="*80)
    print(f"{'Structure':<20} | {'Itérations':>12} | {'Énergie':>15} | {'Erreur moyenne':>15} | {'Temps (s)':>10}")
    print("-" * 80)
    print(f"{'Réseau de câbles':<20} | {result_cables.nit:>12} | {result_cables.fun:>15.6f} | {np.mean(erreurs_cables):>15.6f} | {time_cables:>10.3f}")
    print(f"{'Dôme de tenségrité':<20} | {result_dome.nit:>12} | {result_dome.fun:>15.6f} | {np.mean(erreurs_dome):>15.6f} | {time_dome:>10.3f}")
    print("="*80)

    # Analyse des longueurs
    print("\nAnalyse des longueurs des éléments:")

    print("\nRéseau de câbles:")
    for idx, (id1, id2, l0, K) in enumerate(structure_cables.cables[:4]):
        p1 = structure_cables.nodes[id1]
        p2 = structure_cables.nodes[id2]
        L = np.linalg.norm(p1 - p2)
        print(f"  Câble {id1}-{id2}: L0={l0:.3f}, L={L:.3f}, élongation={(L-l0)/l0*100:.1f}%")

    print("\nDôme de tenségrité:")
    print("  Barres:")
    for idx, (id1, id2, l0, C) in enumerate(structure_dome.bars[:2]):
        p1 = structure_dome.nodes[id1]
        p2 = structure_dome.nodes[id2]
        L = np.linalg.norm(p1 - p2)
        print(f"    Barre {id1}-{id2}: L0={l0:.3f}, L={L:.3f}, élongation={(L-l0)/l0*100:.1f}%")

    print("  Câbles:")
    for idx, (id1, id2, l0, K) in enumerate(structure_dome.cables[:2]):
        p1 = structure_dome.nodes[id1]
        p2 = structure_dome.nodes[id2]
        L = np.linalg.norm(p1 - p2)
        etat = "tendu" if L > l0 else "détendu"
        print(f"    Câble {id1}-{id2}: L0={l0:.3f}, L={L:.3f}, {etat}")

    print("\n✓ Partie III terminée avec succès")
    return True

# ============================================================================
# PROGRAMME PRINCIPAL
# ============================================================================

def main():
    """Fonction principale qui exécute tout le projet"""

    print("\n" + "="*70)
    print("MINI-PROJET D'OPTIMISATION - ENIT 2025/2026")
    print("="*70)

    print("\nMenu principal:")
    print("1. Partie I: Optimisation sans contraintes")
    print("2. Partie II: Optimisation avec contraintes")
    print("3. Partie III: Optimisation des structures de tenségrité")
    print("4. Tout exécuter")
    print("5. Quitter")

    while True:
        choix = input("\nVotre choix (1-5) [4]: ").strip()

        if choix == '':
            choix = '4'

        if choix == '1':
            partie_I()
            break
        elif choix == '2':
            partie_II()
            break
        elif choix == '3':
            partie_III()
            break
        elif choix == '4':
            partie_I()
            partie_II()
            partie_III()
            break
        elif choix == '5':
            print("\nAu revoir!")
            return
        else:
            print("Choix invalide. Veuillez entrer un nombre entre 1 et 5.")

    # Récapitulatif des fichiers générés
    print("\n" + "="*70)
    print("FICHIERS GÉNÉRÉS")
    print("="*70)

    fichiers = [
        "figure1_J2_visualisation.png",
        "figure2_analyse_spectrale.png",
        "figure3_trajectoires_convergence.png",
        "figure4_comparaison_dimensions.png",
        "figure5_solutions_obstacle.png",
        "figure6_gradient_projete.png",
        "figure7_penalisation.png",
        "figure8_uzawa.png",
        "figure9_comparaison_methodes.png",
        "figure10_reseau_cables.png",
        "figure11_dome_tensegrity.png",
        "figure12_comparaison_structures.png"
    ]

    for i, fichier in enumerate(fichiers, 1):
        print(f"{i:2d}. {fichier}")

    print("\n" + "="*70)
    print("PROJET TERMINÉ AVEC SUCCÈS")
    print("="*70)

# ============================================================================
# EXÉCUTION
# ============================================================================

if __name__ == "__main__":
    main()
```
